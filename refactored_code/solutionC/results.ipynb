```python
import plotly.graph_objects as go
import pandas as pd

# Data
tdeltasizes = [5, 10, 15, 20, 25, 30]
data = {
    5: {'fps': 26.32640800888209, 'qgis': 0.3733778993288676, 'Uninterrupted_animation_fps': 13.39125858543665},
    10: {'fps': 23.471724812179776, 'qgis': 0.39918972895695615, 'Uninterrupted_animation_fps': 25.05074473265889},
    15: {'fps': 22.294005466111944, 'qgis': 0.4455266263749864, 'Uninterrupted_animation_fps': 33.66802142005975},
    20: {'fps': 23.960970982093674, 'qgis': 0.4292669636862619, 'Uninterrupted_animation_fps': 46.591053334859915},
    25: {'fps': 24.52577980538303, 'qgis': 0.4490669171015422, 'Uninterrupted_animation_fps': 55.67099033115157},
    30: {'fps': 24.55465709104865, 'qgis': 0.46452908515930175, 'Uninterrupted_animation_fps': 64.58153204704513}
}

# Create a DataFrame from the data
df = pd.DataFrame(data).T
df['tdeltasize'] = tdeltasizes

# Plot 1: Avg fps for each tdeltasize
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=tdeltasizes, y=df['fps'], mode='lines+markers', name='Average FPS'))
fig1.update_layout(title='Average FPS for each TDelta Size', xaxis_title='TDelta Sizes', yaxis_title='Average FPS')

# Plot 2: Avg qgis time in seconds for each tdeltasize
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=tdeltasizes, y=df['qgis'], mode='lines+markers', name='Average QGIS Time'))
fig2.update_layout(title='Average QGIS Time for each TDelta Size', xaxis_title='TDelta Sizes', yaxis_title='Average QGIS Time (seconds)')

# Plot 3: Bar plot comparing avg_fps with Uninterrupted_animation_fps for each tdeltasize
fig3 = go.Figure()
fig3.add_trace(go.Bar(x=tdeltasizes, y=df['fps'], name='Average FPS', width=0.4, offset=-0.2))
fig3.add_trace(go.Bar(x=tdeltasizes, y=df['Uninterrupted_animation_fps'], name='Uninterrupted Animation FPS', width=0.4, offset=0.2))
fig3.update_layout(title='Comparison of Average FPS and Uninterrupted Animation FPS for each TDelta Size',
                   xaxis_title='TDelta Sizes', yaxis_title='FPS', barmode='group')

fig1.show()
fig2.show()
fig3.show()
```
