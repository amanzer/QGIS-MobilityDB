```python
import matplotlib.pyplot as plt

# Plotting the records
plt.plot(onf_times, label='ONF Record')
plt.plot(vat_record, label='VAT Record')
plt.plot(geom_record, label='GEOM Record')

# Adding labels and title
plt.xlabel('Frame')
plt.ylabel('Seconds')
plt.title('Configuration : 5821 boats, Laptop')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
percentages = [0.01, 0.1, 0.5, 1]
laptop_times = [12.525, 16.270, 36.004, 74.700]
desktop_times = [2.456, 4.495, 8.454, 16.236]

# Creating DataFrame
data = {
    'Percentage of Boats': percentages * 2,
    'Time (seconds)': laptop_times + desktop_times,
    'Device': ['Laptop'] * 4 + ['Desktop'] * 4
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

sns.set_palette("colorblind")


sns.barplot(x='Percentage of Boats', y='Time (seconds)', hue='Device', data=df)

# Labels and Title
plt.xlabel('Percentage of Boats')
plt.ylabel('Time (seconds)')
plt.title('Time Taken to Fetch TgeomPoints from the Database')
plt.legend(title='Device')


# Show Plot
plt.show()
```
