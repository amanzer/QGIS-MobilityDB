```python
# Read the pickle file
with open(f"/home/ali/QGIS-MobilityDB/refactored_code/solutionA/laptop_results/Solution_A_1_fps_record.pickle", "rb") as file:
    fps_record = pickle.load(file)

with open(f"/home/ali/QGIS-MobilityDB/refactored_code/solutionA/laptop_results/Solution_A_1_vat_record.pickle", "rb") as file:
    vat_record = pickle.load(file)

with open(f"/home/ali/QGIS-MobilityDB/refactored_code/solutionA/laptop_results/Solution_A_1_geom_record.pickle", "rb") as file:
    geom_record = pickle.load(file)

avg_fps = sum(fps_record) / len(fps_record)

onf_times = [1/fps for fps in fps_record]

avg_onf = sum(onf_times) / len(onf_times)


avg_vat = sum(vat_record) / len(vat_record)
avg_geom = sum(geom_record) / len(geom_record)


print(f"Average FPS: {avg_fps}")
print(f"Average ONF: {avg_onf}")
print(f"Average VAT: {avg_vat}")
print(f"Average GEOM: {avg_geom}")
```

```python
import matplotlib.pyplot as plt

# Plotting the records
plt.plot(onf_times, label='OnNewFrame times')
plt.plot(vat_record, label='ValueAtTimestamp times')
plt.plot(geom_record, label='geometries update time')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Records')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
